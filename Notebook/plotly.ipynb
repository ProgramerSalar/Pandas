```python
# from a pandas dataframe 
batsman = pd.read_csv('C:\\Users\\Mandhata Kumar\\Machine_Learning\\Pandas\\Data\\sharma-kohli.csv')
batsman

plt.plot(batsman['index'], batsman['V Kohli'])
plt.title('runs vs year')
```

```python
# plotting multiple plots 
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# label title 
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scord')
```

```python
# change the color 
# label title 
plt.plot(batsman['index'], batsman['V Kohli'], color="red")
plt.plot(batsman['index'], batsman['RG Sharma'], color='pink')

plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scord')
```

```python
# change the color 
# label title 
plt.plot(batsman['index'], batsman['V Kohli'], color="red",linestyle='solid', linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], color='pink',linestyle='dashdot', linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scord')
```

```python
# change the color 
# label title 
plt.plot(batsman['index'], batsman['V Kohli'], color="red",linestyle='solid', linewidth=3, marker='D', markersize=10)
plt.plot(batsman['index'], batsman['RG Sharma'], color='pink',linestyle='dashdot', linewidth=2, marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scord')
```

```python
# change the color 
# label title 
plt.plot(batsman['index'], batsman['V Kohli'], color="red",linestyle='solid', linewidth=3, marker='D', markersize=10, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='pink',linestyle='dashdot', linewidth=2, marker='o', label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Carrer Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scord')
```

```python
# limiting axes 
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year, price)
plt.ylim(0, 75000)
plt.xlim(2017, 2019)
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
# show function
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
plt.show()
```

### Bar chart

image.png

* Bivariate Analysis 
* Numerical vs Categorical 
* Use case - Aggregate analysis of groups

```python
# simple bar chart 
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors, children, color='black')
```

```python
# horizontal bar chart 
plt.barh(colors, children, color='black')
```

```python
# color and label 
df = pd.read_csv('C:\\Users\\Mandhata Kumar\\Machine_Learning\\Pandas\\Data\\batsman_season_record.csv')
df.head(10)
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2, color='yellow')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color='red')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2, color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

# Pie Chart

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data, labels=subjects)

plt.show()
```

```python
# dataset 
df = pd.read_csv('C:\\Users\\Mandhata Kumar\\Machine_Learning\\Pandas\\Data\\gayle-175.csv')
df.head()
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors 
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', colors=['blue', 'yellow', 'pink', 'cyan', 'brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
